```python
## Fonctions pour descriptions des dataframes
list_dataset = []
list_shape = []
list_variable = []
list_duplicates = []
list_unique = []
list_null = []
list_null_ratio = []
list_type = []

dic_var_info = {'Dataset': list_dataset,
                'Shape': list_shape,
                'Variable': list_variable,
                'Duplicates': list_duplicates,
                'Unique': list_unique,
                'Null': list_null,
                'Null_ratio': list_null_ratio,
                'Type': list_type}


def var_info(df,df_name,var,print_=0,n=5):

    list_dataset.append(df_name)
    list_shape.append(df.shape)
    list_variable.append(var)
    list_duplicates.append(df.duplicated(subset=[var]).sum())
    list_unique.append(df[var].unique().shape[0])
    list_null.append(df[var].isna().sum().sum())
    list_null_ratio.append((df[var].isna().sum().sum())/df.shape[0])
    list_type.append(df[var].dtypes)

    df_data_model = pd.DataFrame(dic_var_info)

    if print_ == 1:
        varInfo = df_data_model.tail(n)
        return varInfo


def all_var_info(df,df_name):

    last_value = df.columns.tolist()[len(df.columns.tolist())-1]
    for i in df.columns.tolist():
        if i != last_value:
            var_info(df,df_name,i)
    return var_info(df,df_name,last_value,1,df.shape[1])
```

Dans un premier temps, on récupère dans une table le nombre de crédits de nos clients ainsi que le montant de ces crédits dans d'autres établissements bancaires à l'aide du dataframe bureau

```python
# Nombre de crédits de chaque client dans d'autres établissements bancaires
previous_credits_count = bureau.groupby('SK_ID_CURR',
                        as_index=False)['SK_ID_BUREAU'].count().rename(columns={'SK_ID_BUREAU': 'Count_previous_credit'})
print(previous_credits_count.shape)
previous_credits_count.head()
```

```python
# Somme totale des crédits de chaque client dans d'autres établissements bancaires
previous_credits_amount = bureau.groupby('SK_ID_CURR',
                        as_index=False)['AMT_CREDIT_SUM'].sum().rename(columns={'AMT_CREDIT_SUM': 'Amount_previous_credit'})
print(previous_credits_amount.shape)
previous_credits_amount.head()
```

```python
# Jointure des deux tables précédentes : table Previous_credits
previous_credits= previous_credits_count.merge(previous_credits_amount, on='SK_ID_CURR', how='left')
previous_credits.head(10)
```

```python
# On ajoute cette nouvelle table au dataframe principal app_train
app_train_bis = app_train.merge(previous_credits, on='SK_ID_CURR', how='left')
app_train_bis.head()
```

```python
app_train_bis["Count_previous_credit"].isna().sum()
```

Pour 44020 clients, il n y a pas de crédits antérieurs dans d'autres établissements bancaires. Nous allons donc remplacer les NAN par 0 pour le nombre de crédits antérieurs et pour le montant des crédits antérieurs

```python
app_train_bis['Count_previous_credit'] = app_train_bis['Count_previous_credit'].fillna(0)
app_train_bis['Amount_previous_credit'] = app_train_bis['Amount_previous_credit'].fillna(0)
app_train_bis.head()
```

### Traitement des valeurs manquantes

```python
app_train_bis.isna().sum()
```

```python
# Pour les colonnes de type "float" et "int", nous utiliserons SimpleImputer pour remplacer les valeurs manquantes par la moyenne
from sklearn.impute import SimpleImputer

col_float = list(app_train_bis.select_dtypes('float64').columns)
col_int = list(app_train_bis.select_dtypes('int64').columns)
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
app_train_bis[col_float] = imp_mean.fit_transform(app_train_bis[col_float])
app_train_bis[col_int] = imp_mean.fit_transform(app_train_bis[col_int])
```

```python
# Pour les autres colonnes , nous remplacerons les valeurs nulles par "Unknown"
col_object=list(app_train_bis.select_dtypes('object').columns)
imp_object = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='Unknown')
app_train_bis[col_object] = imp_object.fit_transform(app_train_bis[col_object])
```

```python
# On verifie qu'il n y a plus de valeur manquante pour app_train_bis
app_train_bis.isna().sum()
```
